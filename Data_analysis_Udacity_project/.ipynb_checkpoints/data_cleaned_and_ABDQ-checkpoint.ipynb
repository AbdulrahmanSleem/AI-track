```python
#importing important files 
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# Load your data and print out a few lines. Perform operations to inspect 
tmdb_data = pd.read_csv('tmdb-movies.csv')
tmdb_data.head()
```

```python
tmdb_data.info()
```

```python
tmdb_data.shape
```

```python
#creating a list of columb to be deleted
tmdb_data.drop(labels = ['id', 'imdb_id', 'budget_adj', 'revenue_adj', 'homepage', 'keywords', 'overview', 'production_companies', 'vote_average'],axis =1,inplace =True)
tmdb_data.head()
```

```python
tmdb_data.dtypes
```

```python
#emoving the duplicate rows if any
tmdb_data.drop_duplicates(inplace=True)
tmdb_data.duplicated().sum()
```

```python
#Removing 0's from budget and the revenue columns
temp_list=['budget', 'revenue']
tmdb_data[temp_list] = tmdb_data[temp_list].replace(0, np.NAN)
tmdb_data.dropna(subset = temp_list, inplace = True)
tmdb_data.describe()
```

```python
tmdb_data.head()
```

```python
#Changing the release date column into standard date format
tmdb_data.release_date = pd.to_datetime(tmdb_data['release_date'])
tmdb_data.head()
```

```python
# Replacing zero with NAN in runtime column
tmdb_data['runtime'] =tmdb_data['runtime'].replace(0, np.NAN)
tmdb_data.dtypes
```

```python
#Changing format of budget and revenue column.
change_type=['budget', 'revenue']
tmdb_data[change_type]=tmdb_data[change_type].applymap(np.int64)
tmdb_data.dtypes
```

```python
tmdb_data.to_csv('tmdb-movies-new.csv', index= False)
```

# Question 4 : What kinds of properties are associated with movies that have high revenues?

we need to discover the property of the revenue

```python
# Discover the statistics of revenue
tmdb_data.revenue.describe()
```

```python
# Draw the histogram of the revenue
tmdb_data.revenue.hist();
```

```python
# Draw the line char for the change of the revenue
revenues = tmdb_data.groupby('release_year')['revenue'].mean()

plt.plot(revenues)
plt.title('The revenue change from year to year')
plt.xlabel('Year')
plt.ylabel('Revenue')

plt.show()
```

 Next, we need to compare the revenue in different years. Since there are too many years, so adding them into decades

```python
#creating decades 
bin_edges = [1960, 1970, 1980, 1990, 2000, 2010, 2015]
bin_names = ['1960s', '1970s', '1980s', '1990s', '2000s', '2010s']
tmdb_data['decade'] = pd.cut(tmdb_data['release_year'], bin_edges, labels=bin_names)
tmdb_data.head()
```

```python
# Draw the bar chart to compare the revenue in differet decades
rev_dis = tmdb_data.groupby('decade').revenue.mean()

plt.bar(rev_dis.index, rev_dis.values)
plt.title('The comparison of revenue among decades')
plt.xlabel('Decade')
plt.ylabel('Revenue')

plt.show()
```

From the bar char above, the revenue increased over the year. The 2010s accounts for the most, while the 1960s accounts for the least.

```python
# we need to find the pattern of the high revenue.
#To order to discover the pattern of the group of high revenues, it is defined that the high revenue is the revenue above the 75% percentile
high_revenue = tmdb_data.revenue.quantile(.75)
high_rev_df = tmdb_data[tmdb_data.revenue > high_revenue]
high_rev_df.revenue.describe()
```

```python
# Draw the scatterplot of the budget and revenue
plt.scatter(x=high_rev_df.budget, y=high_rev_df.revenue)
plt.title('The correlation between Budget and Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')

plt.show()
```

```python
# Calculate the correlation between 'budget' and 'revenue'
high_rev_df[['budget', 'revenue']].corr()
```

The correlation of 0.56 is relatively high for these two variables budget and revenue.

```python
# Draw the scatterplot of the popularity and revenue
plt.scatter(x=high_rev_df.popularity, y=high_rev_df.revenue)
plt.title('The correlation between Popularity and Revenue')
plt.xlabel('Popularity')
plt.ylabel('Revenue')

plt.show()
```

```python
# Calculate the correlation between 'popularity' and 'revenue'
high_rev_df[['popularity', 'revenue']].corr()
```

The correlation between these popularity and revenue is 0.51, a relatively high figure

```python
high_rev_df.apply(pd.to_numeric, errors='coerce').corr()
```

### conclusion  

Overall, for the high revenue group, the revenue is related to the 'popularity' and 'budget' , less related to 'run time' and 'release date'

# 7- Successful genres that have high profit

First, we need to clean this absolute | from genres column

```python
# take a column, and separate the string by '|'
suc_genre = profit_ana['genres'].str.cat(sep = '|')
#giving suc_genre series and storing the values separately
suc_genre = pd.Series(suc_genre.split('|'))
#counting and arranging in descending order
suc_genre = suc_genre.value_counts(ascending = False)

suc_genre
```

```python
# Draw the pie chart of genres
# lt = suc_genre.plot.pie()
sizes = suc_genre.values
labels = suc_genre.index

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')
plt.title('The percentage of each genre')
```

## Conclusion 
from the pie chart and analysis 
Drama is the most profit genre

## 8- Which genres are most popular from year to year?


```python
#storing ['release_year', 'genres', 'popularity'] in a new DataFrame called df_genres:
df_genres = tmdb_data[['release_year', 'genres', 'popularity']]
df_genres
```

```python
# split the genres string
df_genres.genres = tmdb_data.genres.str.split('|')
# # create genre list( create each row for each gen) using explode 
df_genres = df_genres.explode('genres')
df_genres

# groupby year again and get the largest value
data = df_genres.groupby(['release_year','genres'])['popularity'].mean().groupby(level='release_year').nlargest(1).reset_index(level=0, drop=True)
# change the series to dataframe
data = data.reset_index()
data.head(5)
```

```python
# Draw the scatterplot to show the change of the most popular genre
plt.scatter(data.release_year, data.genres)
plt.title('The change of the most popular genre over the year')
plt.xlabel('Year')
plt.ylabel('Genre')

plt.show()
```

## Conclusion
From the scatterplot above, it is easy to show that the most popular genre changed over the year. However, in some periods there are no or little changes. Overall, the most popular genre in 1990s is animation, while it becomes fantasy in 2000s.

## 9- Movies with largest and lowest budgets

```python
# Note: I selected the top 10 movies with the largest budgets
sorted_by_budget = tmdb_data.sort_values('budget', ascending=False)
top_10_budget_movies = sorted_by_budget.head(10)
top_10_budget_movies['budget'].head(10)
```

```python
# Note: I selected the top 10 movies with the largest budgets
sorted_by_budget = tmdb_data.sort_values('original_title', ascending=False)
top_10_budget_movies = sorted_by_budget.head(10)
top_10_budget_movies['budget'].head(10)
```

```python
top_10_budget_movies['budget'].mean()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_10_budget_movies['original_title'], top_10_budget_movies['budget'])
plt.xlabel('Movie Title')
plt.ylabel('Budget (in billions)')
plt.title('Top 10 Movies with the Largest Budgets')
plt.xticks(rotation=90)
plt.show()
```

```python
# Note: I selected the top 10 movies with the lowest budgets
lowest_10_budget_movies = sorted_by_budget.tail(10)
lowest_10_budget_movies['original_title'].head(10)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(lowest_10_budget_movies['original_title'], lowest_10_budget_movies['budget'])
plt.xlabel('Movie Title')
plt.ylabel('Budget (in millions)')
plt.title('Top 10 Movies with the Lowest Budgets')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(top_10_budget_movies['original_title'], top_10_budget_movies['budget'], label='Largest Budget', color='blue')
plt.scatter(lowest_10_budget_movies['original_title'], lowest_10_budget_movies['budget'], label='Lowest Budget', color='red')
plt.xlabel('Movie Title')
plt.ylabel('Budget')
plt.title('Movies with Largest and Lowest Budgets')
plt.legend()
plt.xticks(rotation=90)
plt.show()

# The x-axis will represent the movie titles, and the y-axis will represent the budget.
# The movies with the largest budget in blue and the movies with the lowest budget in red.
```

## 10- Average runtime of all the movies.

```python
average_runtime = tmdb_data['runtime'].mean()
average_runtime
```

```python
plt.figure(figsize=(10, 6))
plt.hist(tmdb_data['runtime'], bins=20, edgecolor='k')
plt.axvline(average_runtime, color='red', linestyle='--', label='Average Runtime')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Runtimes')
plt.legend()
plt.show()

# Note: The red dashed line represents the average runtime
```

```python
avg_runtime = tmdb_data.groupby('release_year')['runtime'].mean().reset_index()
plt.figure(figsize=(12, 6))
plt.scatter(avg_runtime['release_year'], avg_runtime['runtime'])
plt.xlabel('Release Year')
plt.ylabel('Average Runtime')
plt.title('Average Runtime of Movies over the Years')
plt.show()

# This scatter plot will visualize the trend of the average movie runtime over the years
# The release year on the x-axis and the average runtime on the y-axis
```

## Conclusion 
#### In earlier years, movies tended to have shorter runtimes, possibly due to different storytelling and filmmaking conventions.
#### In recent years, movies have generally had longer runtimes, potentially reflecting changes in audience preferences and the ability to accommodate more complex narratives.
